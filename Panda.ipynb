# Concat Dataframes

```python
import pandas as pd
bd_weather=pd.DataFrame({
    'City': ['Dhaka', 'Chattagram', 'Rajshahi',],
    'Temperature': [41, 32, 37],
    'Humidity': [80, 68, 76]
})
bd_weather
```

```python
India_weather=pd.DataFrame({
    'City': ['Delhi', 'Kolkata', 'Rajsthan',],
    'Temperature': [43, 38, 33],
    'Humidity': [77, 82, 65]
})
India_weather
```

```python
df = pd.concat([bd_weather, India_weather])
df
```

```python
df = pd.concat([bd_weather, India_weather], ignore_index=True)
df
```

```python
df = pd.concat([bd_weather, India_weather], keys=["BD", "India"])
df
```

```python
df.loc["BD"]
```

```python
import pandas as pd
temperature_df=pd.DataFrame({
    'City': ['Dhaka', 'Chattagram', 'Rajshahi',],
    'Temperature': [41, 32, 37],
})
temperature_df
```

```python
import pandas as pd
windspeed_df=pd.DataFrame({
    'City': ['Dhaka', 'Chattagram', 'Rajshahi',],
    'windspeed': [7, 12, 9],
})
windspeed_df
```

```python
df = pd.concat([temperature_df, windspeed_df], axis=1)
df
```
