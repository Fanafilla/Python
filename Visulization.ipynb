```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y, color='green', linewidth=5)
```

```python
plt.plot(x,y, 'rD')
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y, color='#008040', marker='+', linestyle='', markersize=20)
```

```python
plt.plot(x,y, color='green', alpha=.2)
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t, label='max')
plt.plot(days,min_t, label='min')
plt.plot(days,avg_t, label='avg')
plt.legend(loc="best", shadow=True, fontsize='large')
plt.grid()
```

# Bar Chart

```python
import numpy as np
Company=['Google', 'AMZN', 'MSFT', 'FB']
Revenue=[90,136,89,27]
Profit=[40,2,34,12]
```

```python
xpos = np.arange(len(Company))
xpos
```

```python
plt.xticks(ypos, Company)
plt.ylabel('Revenue(bln)')
plt.title('BD Tech Stocks')
plt.bar(xpos-0.2, Revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2, Profit, width=0.4, label="Profit")
plt.legend()
```

```python
plt.xticks(ypos, Company)
plt.title('BD Tech Stocks')
plt.barh(xpos-0.2, Revenue, label="Revenue")
plt.barh(xpos+0.2, Profit, label="Profit")
plt.legend()
```

# Histograms

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar_men=[113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women=[67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood sugar analysis')

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 200], rwidth=0.95, color=['green', 'orange'],
         label=['men', 'women'])
plt.legend()
```

# Pi Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
exp_vals=[1400,600,300,410,250]
exp_labels=["Home Rent", "Food", "Phone/Internet bill", "Car", "Other Untilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct="%0.0f%%", shadow=True, explode=[0,0.3,0,0,0], startangle=180)
plt.show()
plt.savefig("piechart.png")
```
